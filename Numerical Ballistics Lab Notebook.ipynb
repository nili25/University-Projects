### 2.1 Theoretical Analysis of the Role of Drag

The initial part of the project is solely based on understanding the mechanics of a dropped cannonball when in a model with air resistance. Key equations
are derived to obtain implicit solutions for integrals. In this instance, wind force is not considered. 
The equation below resolves the force equations of a dropped cannonball when only the weight force and air resistance are taken into account.

**<center>$m\frac{dv_{z}}{dt}=-mg+bv_{z}^{2}\tag{Equation 1}$</center>**

Where $m$ is the mass of the cannonball, $g$ is acceleration, $v_{z}$ is the vertical velocity (in the $z$ axis) and $b$ is the 
drag constant expanded below.

<center>$b=\frac{mg}{u_{T}^{2}}\tag{Equation 2}$</center>

**Equation 1** is integrated in terms of $du$ to obtain an equation that produces a value for time.

<center>$t=\int_0^v \frac{m}{bu^{2}-mg}du\tag{Equation 3}$</center>

Whereby $v$ represents $v_{z}$. When the terms in **Equation 3** are substituted with **Equation 2** (in terms of $u_{T}$), and 
integrated, a final eqation representing the acceleration as a function of time and velocity is derived.

<center>$ gt = \int_0^v \frac{u_{T}^{2}}{u^{2}-u_{T}^{2}}$</center>




Using integration rules with logarithms and natural logs.

<center>$\frac{1}{x} = ln(x)+c$</center>

<center>$gt=\frac{1}{2}u_{T}ln(u_{T}-v_{z})-\frac{1}{2}u_{T}ln(u_{T}+v_{z})\tag{Equation 4}$</center>

An equation that produces a value for the velocity of the dropped
cannonball at time $t$ is derived.

**<center>$tanh(x)=\frac{e^{2x}-1}{e^{2x}+1}\tag{Equation 5}$</center>**

<center>$e^{2\frac{gt}{u_{T}}}=ln\frac{(u_{T}-v_{z})}{(u_{T}+v_{z})}$</center>

Where $a$ replaces the fraction by which $e$ is raised to the power by.

<center>$e^{a}u_{T}-u_{T}=-v_{z}-e^{a}v_{z}$<center>

<center>$-v_{z}=\frac{u_{T}(e^{a}-1)}{(1+e^{a})}$</center>

The substituion of $a$ and trigonometry identities present the following equation.

**<center>$v_{z}=-u_{T}tanh(\frac{gt}{u_{T}})\tag{Equation 6}$</center>**

**Equation 6** is used to plot a graph of velocity against time for a variety of magnitudes of $u_{T}$, when the cannonball is dropped.
The values for $u_{T}$ range from 1 to 6 and increment by 0.2, whilst for $t$, it increments from 0 to 5 $s$ in similar increments.

```python
import numpy as np

u = np.arange (1, 6, 0.2)
t = np.arange(0, 5, 0.2) 

ulist = []

for i in range (0,len(u),1):         #loop increments from 0 to the number of values in the u list
    velocity = []                    #velocity array stores the calculated vz values, a new array for velocity is created for each u_T value
    
    for j in range (0,len(t),1):     #second loop increments as many times as there are variables in the t list
        g = 10
        vz = -u[i] * np.tanh((g*t[j])/u[i])      #equation 6 written out in terms of code to calculate the velocity at a certain time value
        velocity.append(vz)
            
    ulist.append(velocity)
```

```python
import matplotlib.pyplot as plt
for i in range (0,len(t),1):
    plt.plot(t,ulist[i])             #plotting of the graph for every value of t with respect to vz

plt.xlabel('time / s')
plt.ylabel('velocity / $ms^{-1}$')
plt.title('Velocity of the dropped cannonball as a function of Time')
plt.show()
```

<b>Model affected by air resistance:</b>

**Equation 6** is integrated to produce the equation for height of the cannonball at time $t$, as seen below.

<center>$u_{T}\int tanh(\frac{gt}{u_{T}}) = lncosh(\frac{gt}{u_{T}})\frac{u_{T}}{g}+C$</center>

When integrating the following equation, the $+C$ is representative of the $z_{0}$ as the velocity, when integrated, reverts to displacement.

<center>$z=z_{0}-\frac{u_{T}^{2}}{g}lncosh(\frac{gt}{u_{T}})\tag{Equation 7}$</centerr>

<b>Equation 7</b> is used as opposed to general suvat equations since the air resistance is now considered. 

```python
import math
import numpy as np

height = np.arange(0,100,1)
time = []                     #time array created to store the respective values of t for each height variable

for i in range (0, 100, 1):
  num = -i/1000
  var = (math.e**(-num))
  inv = math.acosh(var)
  t = inv * 10                #equation 6 rearranged and simplified to code easily
  time.append(t)

import matplotlib.pyplot as plt
for i in range (0,len(time),1):
    plt.plot(time,height)         

plt.xlabel('time / s')
plt.ylabel('height / $m$')
plt.title('Height against time with air resistance')
plt.show()

num = 1/10                   #1/10 is the final value for num, so the calculation for t is given below to show the total time taken
var = (math.e**(num))
inv = math.acosh(var)
t = inv *10

print ("The time taken when air resistance is present is:", t, "s")
```

### 2.2 Numerical study of the Role of Drag

The model of the cannonball affected by air resistance is further expanded to account for a 2 dimensional ($x$,$z$) plane. In this study, the cannonball is now launched as opposed to being dropped from a certain height. 
$v_{y}$ is set to 0 to focus on only the 2 planar axis. The equations used to solve the change in velocity and 
acceleration for both the $x$ and $z$ plane are seen below.

**<center>$m\frac{dv_{x}(t)}{dt}=-\frac{mg}{u_{T}^{2}}\sqrt{v_{x}(t)^{2}+v_{z}(t)^{2}}v_{x}(t)\tag{Equation 8}$</center>**


**Equation 8** is representative of the acceleration in the $x$ axis only 

**<center>$m\frac{dv_{z}(t)}{dt}=-\frac{mg}{u_{T}^{2}}\sqrt{v_{x}(t)^{2}+v_{z}(t)^{2}}v_{z}(t)-mg\tag{Equation 9}$</center>**

**Equation 9** is representative of the accleration in the $z$ axis only

The table below shows the initial conditions for the variables in the equations

| vx0 / ms^-1 | vz0 / ms^-1 | z0 / m    | x0 / m   | uT / ms^-1  | g / ms^-2
|-------------|-------------|-----------|----------|-------------|-------------
|500          |500          |0          |0         |100          |10

<center>Table 2.0 Table showing the values for the initial speeds and displacements for both $x$ and $z$ planes, along with the terminal velocity and graviational acceleration</center>

The three graphs below show the variation of certain variables whilst the cannonball is airborne. The time ranges from 0 to 50 $s$ in increments of 0.5 $s$. To solve **Equations 8** and **9**, a recursive function algorithm is appended to the code to allow all possible integrals to be solved. 

<b>Height of projectile as a function of time</b>

```python
from scipy.integrate import odeint
import numpy as np
import math

g = 10             
u = 100             
bvalue = g/(u**2)        # for ease of code the variables made of other variables are written out as one variable

x0=0               
z0=0
vx0 = 500
vz0 = 500

## define function to compute f(X,t)
def f_func(w,t):
    
    f = np.zeros(4)
    x,z,vx,vz = w
    
    f[0] = w[2]                                                    # f[0] = x component of velocity
    f[1] = w[3]                                                    # f[1] = z component of velocity
    f[2] = - bvalue * (f[0]**2 + f[1]**2)**(0.5) * f[0]            # f[2] = acceleration in x direction
    f[3] = - bvalue * (f[0]**2 + f[1]**2)**(0.5) * f[1] - g        # f[3] = acceleration in z direction
    return f

w0 = [x0, z0, vx0, vz0]        # set initial state of the system

t = np.arange(0, 50, 0.5)  

## solve ODE using odeint
solution = odeint(f_func,w0,t) 

import matplotlib.pyplot as plt

x = solution[:,0]             # split up the returned solution matrix that consists of multiple columns of values that represent different data values
z = solution[:,1] 
vx = solution[:,2] 
vz = solution[:,3]
```

```python
plt.plot(t,z)
plt.xlabel('time / $s$')
plt.ylabel('$z$ / $m$')
plt.ylim(0, None)            # set a limit to how much the program plots on the graph
plt.title('The height, $z$, of the projectile as a function of time')
plt.show()
```

### 2.3 Predict the Landing Point of a Cannonball Without Wind

The following study treats the muzzle velocity ($v_{m}$) of the cannon as fixed. The only changing variable is the direction
the cannon faces (which is affected by the angle at which the cannon points). 

The table below shows the initial condition of the variables. 

| vm / ms^-1 | vy0 / ms^-1 | uT / ms^-1  | g / ms^-2 | z0 / ms |
|------------|-------------|-------------|-----------|---------|
|100         |0            |100          |10         |0        |

<center>Table 3.0 Table showing the initial conditions of the variables such as $v_{m}$, $v_{y0}$, the gravitational acceleration and initial displacements in the vertical axis</center>

The angle of inclination, $\theta$, of the cannon ranges from 5 - 90 degrees in increments of 5 degrees. The time varies from 0 to 17 
$s$ in singular increments. The displacement of the trajectory, $x_{land}$ is observed based in the variations of $\theta$.

```python
from scipy.integrate import odeint
import numpy as np
import math

## set initial conditions and parameters

for i in range (5,95,5):

    g = 10              
    vm = 100            
    u = 100             
    bvalue = g/(u**2)
    x0=0                
    z0=0
    
    th = i * math.pi/180    # convert launch angle to radians
    vx0 = vm*np.cos(th)
    vz0 = vm*np.sin(th)
        
        
    # since th is using an array, and is being split per index to turn into its respective angle in the vx form, 
    # it needs to be split this way as an array can not be passed into the equations below for vx0 and vz0. 

    ## define function to compute f(X,t)
    def f_func(w,t):

        f = np.zeros(4)
        x,z,vx,vz = w

        f[0] = w[2] # f[0] = x component of velocity
        f[1] = w[3] # f[1] = z component of velocity
        f[2] = - bvalue * (f[0]**2 + f[1]**2)**(0.5) * f[0]         # f[2] = acceleration in x direction
        f[3] = - bvalue * (f[0]**2 + f[1]**2)**(0.5) * f[1] - g        # f[3] = acceleration in z direction
        return f

    ## set initial state vector and time array
    w0 = [x0, z0, vx0, vz0]        # set initial state of the system
    t = np.arange(0, 17, 1)  

    ## solve ODE using odeint
    solution = odeint(f_func,w0,t) 

    import matplotlib.pyplot as plt

    x = solution[:,0]  
    z = solution[:,1] 
    vx = solution[:,2] 
    vz = solution[:,3]

    th = np.arange (5,90,5)
    plt.plot(th,x) 
    
plt.xlabel('angle of inclination / degrees')
plt.ylabel('$x_{land}$ / m')
plt.ylim(0, None)
plt.title('Displacement of the cannonball, $x$land, as it varies with Angle of Inclination')
plt.show()
```

To further advance this study, a target is placed a distance of 100 $m$ from the cannon. The muzzle velocity $v_{m}$ of the cannon is kept the same (100 $ms^{-1}$) and the time now varies from 0 to 90 $s$ in increments of 0.01 $s$. The angle of inclination required to hit the target is now determined.

```python
from scipy.integrate import odeint
import numpy as np
import math

## set initial conditions and parameters

for i in range (5,95,5):

  g = 10              
  vm = 100            
  u = 100             
  bvalue = g/(u**2)
  x0=0                
  z0=0

  # since th is using an array, and is being split per index to turn into its respective angle in the vx form, 
  # it needs to be split this way as an array can not be passed into the equations below for vx0 and vz0. 

  for i in np.arange (0,90,0.01):
      th = i * math.pi/180    # convert launch angle to radians
      vx0 = vm*np.sin(th)
      vz0 = vm*np.cos(th)

  ## define function to compute f(X,t)
  def f_func(w,t):

      x,z,vx,vz = w
      f = [vx,- bvalue * (vx**2 + vz**2)**(0.5) * vx,vz,- bvalue * (vx**2 + vz**2)**(0.5) * vz - g]
      
      return f

  ## set initial state vector and time array
  w0 = [x0, z0, vx0, vz0]        # set initial state of the system
  t = np.arange(0, 90, 0.01)  

  ## solve ODE using odeint
  solution = odeint(f_func,w0,t) 

  import matplotlib.pyplot as plt

  x = solution[:,0]  
  z = solution[:,1] 
  vx = solution[:,2] 
  vz = solution[:,3]

  th = np.arange(0,90,0.01)

  plt.plot(th,x)
  
plt.xlabel('angle of inclination / degrees')
plt.ylabel('displacement / m')
plt.title('Displacement of the cannonball, $x$land, as it varies with Angle of Inclination')
plt.xlim(0,10)
plt.ylim(0,400)
plt.show()

land2 = np.where(x<100)          # calculates the angle where the cannonball actually hits or is close to hitting the target
above = th[land2]
last = above[len(above)-1]
```

### 2.4 Direct the Landing Point of a Cannonball With Wind

The main objective of the project is now examined. In this system, all 3 dimensions are acknowledged, which 
therefore results in both the angle of inclination, $θ$, and the azimuthal angle, $ϕ$, considered. 
In order to solve the landing point for the cannonball given the 3 planes are considered, an extension to **Equations 7** and **8** are now solved to account for the wind speeds. 

**<center>$ma=F_{grav}+F{wind}\tag{Equation 10}$</center>**

**<center>$m\frac{d}{dt}\begin{pmatrix} v_{x}(t) \\ v_{y}(t) \\ v_{z}(t) \end{pmatrix}=-mg \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix} - b \sqrt{(v_{x}(t)-w_{x})^{2}+(v_{y}(t)-w_{y})^{2}+v_{z}^{2}}\begin{pmatrix} v_{x}(t)-w_{x} \\ v_{y}(t)-w_{y} \\ v_{z}(t) \end{pmatrix} \tag{Equation 11}$</center>**

The velocities in the $x$ and $y$ plane are being subtracted from the wind speed as the effects of wind result in a new velocity for the different dimensions. 

When coded, the equation is split into three to include the vector quantities of the planar values. The target is placed at 100 $m$ as in the previous study. 

The table below shows the initial conditions of the variables.

| wx / ms^-1 | wy / ms^-1 | xhit / m  | yhit / m |
|------------|------------|-----------|----------|
|10          |0           |100        |0         |

<center>Table 4.0 Table showing the intial conditions of variables, such as $w_{x}$ and $x_{hit}$, used in the equation</center>

The time ranges for this system is from 0 to 150 $s$ in increments of 0.04 $s$. 

```python
from scipy.integrate import odeint
import numpy as np
import math

## set initial conditions and parameters

g = 10
u = 100
bvalue = -g/(u**2)
vm = 100
wx = 10
wy = 0
x0=0 
y0=0
z0=0

t = np.arange(0, 150, 0.04)  

#function has been defined with the index positions of f array containing the equations used for integration
def f_func(w,t):
    
    x,vx,y,vy,z,vz = w
    f = [vx, (bvalue * ((vx-wx)**2 + (vy-wy)**2 + vz**2)**(0.5)) * (vx-wx), vy, (bvalue * ((vx-wx)**2 + (vy-wy)**2 + (vz)**2)**(0.5)) * (vy-wy) ,vz, -g + (bvalue * ((vx-wx)**2 + (vy-wy)**2 + (vz)**2)**(0.5)) * (vz)]
    return f


for j in np.arange (0,90,1):
    ph = j * math.pi/180
    for i in np.arange (0,90,1):
        th = i * math.pi/180   # convert launch angle to radians
        vx0 = vm*(np.cos(th))*(np.cos(ph))
        vy0 = vm*(np.cos(th))*(np.sin(ph))
        vz0 = vm*np.sin(th)
        
        w0 = [x0, vx0, y0, vy0,z0, vz0]        # set initial state of the system
        
        solution = odeint(f_func,w0,t) 
        
        x = solution[:,0]          #splitting up of the solution matrix is done here where each column is a different type of data value
        vx = solution[:,1]
        y = solution[:,2] 
        vy = solution[:,3] 
        z = solution[:,4]
        vz = solution[:,5]

        if (y[i] <= 0.1) and ((x[i] >= 99.25) and (x[i] <= 100)):     #checks are made to ensure that the ball has reached the ground and also within the target point
            theta = ((180/math.pi)*th)
            phi = ((180/math.pi)*ph)
            print ("When the y displacement is", y[i], "m", ",the x displacement is", x[i], "m", ",which correlates to theta being", round(theta), "degrees", "and phi being", round(phi), "degrees.")
```

The given $x$ displacement from the program is very close to the actual target that is 100 $m$ away.

```python
from scipy.integrate import odeint
import numpy as np
import math

## set initial conditions and parameters

g = 10
u = 100
bvalue = -g/(u**2)
vm = 100
wx = 10
wy = 0
x0=0 
y0=0
z0=0

    
t = np.arange(0, 150, 0.04)  

def f_func(w,t):

    x,vx,y,vy,z,vz = w
    f = [vx, (bvalue * ((vx-wx)**2 + (vy-wy)**2 + vz**2)**(0.5)) * (vx-wx), vy, (bvalue * ((vx-wx)**2 + (vy-wy)**2 + (vz)**2)**(0.5)) * (vy-wy) ,vz, -g + (bvalue * ((vx-wx)**2 + (vy-wy)**2 + (vz)**2)**(0.5)) * (vz)]
    return f

print ("The impact point of the cannonball with a 10% muzzle velocity error changes as seen below:")

minvm = vm - ((10/100)*vm)        #bounds for the errors are created here for a lower and upper 10 percent
maxvm = vm + ((10/100)*vm)

vm = minvm

while (vm >= minvm) and (vm <= maxvm):
    if vm != 100:
        print ("For a 10% error value of:", vm, "ms^-1")
    for j in np.arange (0,90,1):
        ph = j * math.pi/180
        for i in np.arange (0,90,1):
            th = i * math.pi/180   # convert launch angle to radians
            vx0 = vm*(np.cos(th))*(np.cos(ph))
            vy0 = vm*(np.cos(th))*(np.sin(ph))
            vz0 = vm*np.sin(th)

            w0 = [x0, vx0, y0, vy0,z0, vz0]        # set initial state of the system

            solution = odeint(f_func,w0,t) 

            x = solution[:,0] 
            vx = solution[:,1]
            y = solution[:,2] 
            vy = solution[:,3] 
            z = solution[:,4]
            vz = solution[:,5]

            if (y[i] <= 0.1) and ((x[i] >= 99) and (x[i] <= 100)):         
                theta = ((180/math.pi)*th)
                phi = ((180/math.pi)*ph)
                if vm != 100:
                    print ("The impact point is", x[i], "m") 
    vm = vm + 10       #increments by 10 due to the 10 percent error
```
