In this assignment you will choose an ordinary differential equation (ODE) with a well known solution or one that you know how to solve. As an example, let's choose $y' = \frac{3x^2}{y}$. This has a true solution, which we can find by integration if we have some starting conditions (let's choose $y=1$ when $x=0$). This is a separable differential equation and so rearranging and integrating gives : 

$$ \int{y} dy = \int{3x^2} dx $$
$$ \frac{y^2}{2} = x^3 + c $$
hence, for $y(0) = 1$
$$ c = \frac{1}{2} $$

and hence

$$ y^2 = 2x^3 + 1 $$

therefore the 'real' solution is

$$ y = \sqrt{2x^3 + 1} $$

(taking positive solutions only)

## Part 1 : 
(7 marks out 20). 
Once you have chosen your ODE, write a code to work out a numerical solution to this differential equation. The user should provide as inputs the step-size $h$ and the value of x for which the solution is sought. 

```python
import math
def func(x,y):                      #function that returns the ODE to solve
    return ((3*(x**2))/y)          

step_h = float(input("Please enter in the value of your required step size h: "))
sol_x = int(input("Please enter the finding solution value of x: "))
steps = int(sol_x/step_h)           #the for loop can't iterate through a float number so it has to be changed to an integer

x = 0                               # Set the starting value of x 
y = 1                               # Set the starting value of y 

for i in range(1,steps+1):          # Iterates round the loop "steps" times
    y = y + step_h*func(x,y)        # increments the value of y with the Euler method
    x = x + step_h                  # increments the value of x with step_h

print ("For the x value of: ", x, "the numerical solution of the ODE is: ", y)
```

## Part 2 :

(7 marks out 20). Copy your code from part 1 into the cell below and modify it as follows. 

Add to your code so that it works out the difference between the approximated value and the true value at each step and store this difference as a list. Examine some of the values in this list in your own way, to help you understand what is happening in your approximation. Or you could plot it as a function of x (see example in the Lab3 Notebook).
```python
def func(x,y):                      #function that returns the ODE to solve
    return ((3*(x**2))/y)          

def approximate(sol_x,y):           #function that checks the difference between the two values
    x = sol_x
    true_answer = math.sqrt((2*(x**3))+1)
    difference = true_answer - y
    return difference

step_h = float(input("Please enter in the value of your required step size h: "))
sol_x = int(input("Please enter the finding solution value of x: "))
steps = int(sol_x/step_h)           #the for loop can't iterate through a float number so it has to be changed to an integer

x = 0                               #Set the starting value of x 
y = 1                               #Set the starting value of y 
ydif = []
xdif = []

for i in range(1,steps+1):          #Iterates round the loop "steps" times
    y = y + step_h*func(x,y)        #increments the value of y with the Euler method
    x = x + step_h                  #increments the value of x with step_h
    xdif.append(x)
    dif = approximate(sol_x,y)      #a variable catches the return value from the function approximate
    ydif.append(dif)
   
print ("For the x value of: ", x, "the numerical solution of the ODE is: ", y)

import matplotlib.pyplot as plt       
plt.rcParams['figure.dpi'] = 150      
plt.plot(xdif, ydif, 'b')            
plt.show()                           #Displays the graph
```
